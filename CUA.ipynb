```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.cluster.vq import kmeans2
from numpy.linalg import *
```

```python
from sklearn.datasets import load_iris
data = load_iris()
X = data.data
y = data.target
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
for i in range(3):
    print(f'For i = {i}:')
    new_X = X_test[y_test==i]
    plt.plot(new_X[:,2], new_X[:,1], 'o', c=np.random.random(3))
    
```

```python
def compute_mus_sigmas(Z,J):
    mu, idx = kmeans2(Z,J)
    sigma = np.zeros(J)
    for j in range(J):
        delta = sorted(norm(mu-mu[j], axis=1))
        sigma[j] = (delta[1] + delta[2])/2.0
    return mu, sigma
        
                    
    
```

```python
J = 5
mu, sigma = compute_mus_sigmas(X_train, J)
```

```python
def rbfnn(J, Z, t):
    P = len(Z)
    print(P)
    
    phi = np.zeros((P,J))
    print(phi)
    mu, sigma = compute_mus_sigmas(Z, J)
    for p in range(P):
        for j in range(J):
            phi[p,j] = np.e**(-1*(norm(Z[p]-mu[j])**2)/(2*sigma[j]*sigma[j]))
    print(phi)
                                 
    
    
```
